```python
s1 = pandas.read_csv("../soubory_k_praci/studenti1.csv").dropna()
s2 = pandas.read_csv("../soubory_k_praci/studenti2.csv").dropna()
```

```python
prezencni_stud = pandas.concat([s1, s2], ignore_index=True)
prezencni_stud
```

```python

prezencni_obory = prezencni_stud.groupby('obor')['jméno'].count()
# prezencni_obory = prezencni_stud.groupby('obor').size() # -> započítává NaN hodnoty
prezencni_obory # -> nebo zde přidám ['jméno']
# prezencni_obory
```

```python

prezencni_obory = prezencni_stud.groupby('obor')['prospěch'].mean()
# prezencni_obory = prezencni_stud.groupby('obor').agg({'prospěch': ['count', 'mean', 'median', 'max', 'min']})
# prezencni_obory = prezencni_stud.groupby('obor')['prospěch'].describe() # -> předefinovaný typ statistiky
prezencni_obory
```

```python

prezencni_stud = pandas.merge(prezencni_stud, jmena)
prezencni_stud
```

```python

pohlavi = prezencni_stud.groupby('pohlaví')['jméno'].count()
pohlavi
```
